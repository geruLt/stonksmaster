```python
import torch
import pandas as pd
import numpy as np
import glob

from collections import Counter
```

```python
# Get paths of stonks data from the local database
stock_data = glob.glob('data/*.csv')

# Create holders for x and y
X = []
y = []

features = ['Open', 'High', 'Low', 'Close', 'Adj Close'] # Percent change augmentation columns
interval = 60 # Minutes per episode

counter = 0
for stock_path in stock_data:
    print(stock_path)
    # Read next stock file
    df = pd.read_csv(stock_path)
    df.Datetime = pd.to_datetime(df.Datetime)
    hard_limit = len(df)

    # Get samples from the file
    counter = 0
    while counter < hard_limit-interval:
        start_time = df.iloc[counter].Datetime
        end_time = df.iloc[counter+interval].Datetime
        if (end_time-start_time).seconds == (interval*60): # Meaning no missing data
            new_df = df.drop(['Datetime'], axis=1).iloc[counter:counter+interval+1]
            new_df[features] = new_df[features].pct_change().fillna(0)
            X.append(new_df.iloc[:interval].to_numpy())
            y.append(new_df.iloc[interval].to_numpy())
            counter += interval
        else:
            counter += 1
```

```python
'''
Labels:  0: Stock will lose money
         1: Stock will not make money
         2: Stock will make money
'''
y = [2 if v[3] > 0.001 else 1 if v[3] > 0 else 0 for v in y]
Counter(y)
```

```python
# Save samples and labels
X = np.array(X)
y = np.array(y)
y_tensor = torch.from_numpy(y)
onehot_y = torch.nn.functional.one_hot(y_tensor.to(torch.int64), 3)

np.save('data/X.npy', X)
np.save('data/y.npy', onehot_y.numpy())
```
