```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/TSLA.csv')
df.Datetime = pd.to_datetime(df.Datetime)
```

```python
df.head()
```

```python
x = []
y = []
features = ['Open', 'High', 'Low', 'Close', 'Adj Close'] # Percent change augmentation columns
interval = 30 # Minutes per episode
hard_limit = len(df) 
counter = 0
while counter < hard_limit-interval:
    start_time = df.iloc[counter].Datetime
    end_time = df.iloc[counter+interval].Datetime
    if (end_time-start_time).seconds == 1800: # Meaning no missing data
        new_df = df.drop(['Datetime'], axis=1).iloc[counter:counter+interval+1]
        new_df[features] = new_df[features].pct_change().fillna(0)
        x.append(new_df.iloc[:interval].to_numpy())
        y.append(new_df.iloc[interval].to_numpy())
        counter += 30
    else:
        counter += 1
```
